```python
import random 
#creating the tuples and the dictionary for suits and ranks and value of ranks in integer.
suits = ('Hearts' , 'Diamonds', 'Spades' , 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10, 'Queen':10, 'King':10, 'Ace':11}
#creating a boolean value to control the while loop.
playing = True
```

```python
#creating each card and printing the statements.
class Card:
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        
    def __str__(self):
        return f"{self.rank} of {self.suit}"
```

```python
#creating all the 52 cards.
class Deck:
    def __init__(self):
        self.deck = []  #Taking an empty list.
        for suit in suits:
            for rank in ranks:
                self.deck.append(Card(suit,rank)) #Appending each card created (in Cards class) to the empty list.

    def __str__(self):
        a_deck = ''
        for card in self.deck:
            a_deck += '\n ' + str(card)
        return f"The deck has:{a_deck}"#printing each in a new line

    def shuffle(self):
        random.shuffle(self.deck)#shuffling all the cards in the deck.

    def deal(self):
        dealed_one = self.deck.pop()
        return dealed_one #removing the last card from the deck.
        
        
```

```python
class Hand:
    def __init__(self):
        self.cards = []
        self.value = 0 #for knowing the values of the start with zero
        self.aces = 0 #to keep the track of aces

    def add_cards(self, a_card):
        self.cards.append(a_card)
        self.value += values[a_card.rank] #adding the values of the cards
        if a_card.rank == 'Ace':
            self.aces += 1 #if the card is an ace, increment the ace counter

    def adjusting_for_ace(self):
        #Adjust for aces if value exceeds 21.
        while self.value>21 and self.aces>0: 
            self.value -= 10 #reducing the total value by 10(Ace is worth 1 instead of 11).
            self.aces -= 1 #since an ace is adjusted so we are reducing the count of the aces.
       
```

```python
class Chips:
    def __init__(self):
        self.total = 100 #considering some default value.let us consider we have 100 chips.
        self.bet = 0 

    def winning_bet(self):
        self.total += self.bet #after winning the taking all the chips.

    def losing_bet(self):
        self.total -= self.bet #after losing a bet then removing some chips from the total.
```

```python
def taking_a_bet(chips):
    while True:
        try:
            chips.bet = int(input('How many chips would you like to bet? '))
        except ValueError:
            print('Sorry,a bet must be an integer!')
        else:
            if chips.bet > chips.total:
                print("Sorry, your bet can't exceed",chips.total)
            else:
                break
```

```python
def hit(deck,hand):
    acard = deck.deal() #taking a card from the deck
    hand.add_cards(acard) #adding the cards to the hand
    hand.adjusting_for_ace() #adjusting the aces
```

```python
def hit_or_stand(deck,hand):
    global playing
    while True:
        option = input("Enter hit or stand: 'h' or 's' ")
        if option[0].lower() == 'h': #if the option selected was 'h' then hit
            hit(deck,hand)
        elif option[0].lower() == 's': #otherwise stand
            print("Player stands. Dealer is playing.")
            playing = False
            break
        else:
            print("Sorry, please try again.")
            continue
        break
```

```python
def show_some(player,dealer):
    print("Dealer's hand:")
    print("<card hidden>") #hidding a dealer's card
    print('',dealer.cards[1])#printing the other card of the dealer.
    print("\nPlayer's hand:\n",*player.cards,sep = '\n ') #printing all the cards of the player.
    
```

```python
def show_all(player,dealer):
    print("\nDealer's hand:",*dealer.cards,sep = '\n') #printing all the dealer's cards
    print("Dealer's hand value = ",dealer.value) #printing the dealer's cards values
    print("\nPlayer's hand:",*player.cards,sep = '\n') #printing all the player's cards
    print("Player's hand value = ",player.value) #printing the values of the player's cards
```

```python
def player_busts(player,dealer,chips):
    print("Player busts!")
    chips.losing_bet()

def dealer_busts(player,dealer,chips):
    print("Dealer busts!")
    chips.winning_bet()

def player_win(player,dealer,chips):
    print("Player wins!")
    chips.winning_bet()

def dealer_win(player,dealer,chips):
    print("Dealer wins!")
    chips.losing_bet()

def push(player,dealer,chips):
    print("Dealer and Player tie!It's a push.")
```

```python
while True:
    #print an opening statement
    print("Welcome to BlackJack!\n")
    print('RULES:\n -Get as close to 21 as you can without going over.')
    print(' -Dealer hits until he/she reaches 17.\n -Aces count as 1 or 11.\n -Jack,Queen,King counts as 10.\n')

    #creating a deck and shuffling them.
    deck = Deck()
    deck.shuffle()

    #handing over two cards each to the player and the dealer.
    player_hand = Hand()
    player_hand.add_cards(deck.deal())
    player_hand.add_cards(deck.deal())

    dealer_hand = Hand()
    dealer_hand.add_cards(deck.deal())
    dealer_hand.add_cards(deck.deal())
    
    #setting up the player's chips.
    player_chips = Chips() #remember the default value is 100.
    taking_a_bet(player_chips)

    show_some(player_hand,dealer_hand)

    while playing: #from the class hit_or_stand
        hit_or_stand(deck,player_hand) #for plaer to hit or stand
        show_some(player_hand,dealer_hand) #all player cards are shown but a dealer's card is hidden
        #if player exceeds the value 21 then run player_busts and break out of the loop
        if player_hand.value > 21:
            player_busts(player_hand,dealer_hand,player_chips)
            break

    #if player's cards value doesnot exceed 21 then:
    if player_hand.value <= 21:
        #as according to the rules if the dealer's cards values is <17 then he hits
        #dealer's hand plays until the dealer reaches 17
        while dealer_hand.value < 17:
            hit(deck,dealer_hand)
           
        #showing all the cards of the player and dealer along with the values of their cards.
        show_all(player_hand,dealer_hand)

        if dealer_hand.value > 21: #if dealer exceeds 21 he gets busted
            dealer_busts(player_hand,dealer_hand,player_chips)

        elif player_hand.value > dealer_hand.value:
            player_win(player_hand,dealer_hand,player_chips)

        elif player_hand.value < dealer_hand.value:
            dealer_win(player_hand,dealer_hand,player_chips)

        else:
            push(player_hand,dealer_hand,player_chips)

    print("\nPlayer's winnings are ",player_chips.total) #infor player of their chips total.

    new_game = input("Would you like to play a new game? enter 'y' or 'n'")
    if new_game[0].lower == 'y':
        playing = True
        continue

    else:
        print("Thanks for playing the game.")
        break
        
```
